```python
import pandas as pd
import numpy as np
import openpyxl

Departamentos={'Antioquia','Atlántico','Bogotá','Bolívar','Boyacá','Caldas','Caquetá','Cauca','Cesar','Córdoba','Cundinamarca','Chocó','Huila','La Guajira','Magdalena','Meta','Nariño','Norte de Santander','Quindio','Risaralda','Santander','Sucre','Tolima','Valle','Arauca','Casanare','Putumayo','San Andrés y Providencia','Amazonas','Guainía','Guaviare','Vaupés','Vichada'}
tabla1=pd.read_excel("C:/Users/jijapago/Downloads/2007def.xls")


contador=2

while True:

    try:

        if not(tabla1.iloc[contador,0] in Departamentos) and tabla1.iloc[contador,0]!='TOTAL' :

            tabla1=tabla1.drop(contador,axis=0)
            tabla1.reset_index(inplace=True,drop=True)

        elif tabla1.iloc[contador,0]=='TOTAL':

            tabla1.iloc[contador,0]=tabla1.iloc[contador-1,0]

        else: 

            contador+=1

    except IndexError:

        break

contador=1

while True:

    try:

        if tabla1.iloc[contador,0] in Departamentos:

            Departamentos.discard(tabla1.iloc[contador,0])
            contador+=1

        else: 
               
            tabla1=tabla1.drop(contador-1,axis=0)
            tabla1.reset_index(inplace=True,drop=True)


    except IndexError:

        break

tabla_final=tabla1

tabla_final.to_excel('D:/Colombia/Defunciones no fetales/2007.xlsx',index=False)
```

```python
import pandas as pd
import numpy as np

lista=['1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']
for i in lista:

    tabla_aux=pd.read_csv('D:/Colombia/Defunciones no fetales/'+i+'.csv',sep=';',header=None, encoding='latin-1')
    serie1=np.full(33,'HOMBRE')
    serie1=pd.Series(serie1)

    serie2=np.full(33,'MUJER')
    serie2=pd.Series(serie2)

    serie3=np.full(33,'INDETERMINADO')
    serie3=pd.Series(serie3)


    tabla_aux1=tabla_aux.iloc[1:,[0,1]]
    tabla_aux1.reset_index(inplace=True,drop=True)

            
    tabla_aux1=pd.concat([tabla_aux1,serie1],axis=1)

    tabla_aux2=tabla_aux.iloc[1:,[0,2]]
    tabla_aux2.reset_index(inplace=True,drop=True)

            
    tabla_aux2=pd.concat([tabla_aux2,serie2],axis=1)
    tabla_aux2.rename(columns={2:1},inplace=True)

    tabla_aux3=tabla_aux.iloc[1:,[0,3]]
    tabla_aux3.reset_index(inplace=True,drop=True)

            
    tabla_aux3=pd.concat([tabla_aux3,serie3],axis=1)
    tabla_aux3.rename(columns={3:1},inplace=True)

    tabla=pd.concat([tabla_aux1,tabla_aux2,tabla_aux3],axis=0)

    tabla.to_csv('D:/Colombia/Defunciones no fetales/D'+i+'.csv',index=False,sep=';',header=False)



```

```python
import pandas as pd
import numpy as np
import openpyxl

lista=['Departamento 2001','Departamento 2002','Departamento 2003','Departamento 2004 ','Departamento 2005','Departamento 2006']

tabla=pd.DataFrame({'AÑO':[],'Departamentos':[],'Población total':[],'Población en edad de trabajar':[],'Población económicamente activa':[]})



for i in lista:


    tabla_aux=pd.read_excel('D:/Colombia/Empleo/Empleo2001_2006.xls',sheet_name=i)
    serie=pd.Series(np.full(tabla_aux.shape[0],i[-4:]),name='AÑO')
    
    tabla_aux=pd.concat([serie,tabla_aux],axis=1)

    tabla= pd.concat([tabla,tabla_aux],axis=0)

tabla=tabla.iloc[:,0:5]
tabla.dropna(inplace=True)
tabla[['Población total','Población en edad de trabajar','Población económicamente activa']]=tabla[['Población total','Población en edad de trabajar','Población económicamente activa']].replace('.','',regex=True)

print(tabla.head(5))

tabla.to_csv('D:/Colombia/Empleo/tabla1.csv',index=False,sep=';')
```

```python
import pandas as pd
import numpy as np

tabla1=pd.read_csv('D:/Colombia/Empleo/Empleo2007_2022.csv',sep=';',decimal=',',skiprows=1,encoding='latin-1',header=None)

tabla=pd.DataFrame({'Departamento':[],'Año':[],'Población total':[],'Población en edad de trabajar':[],'Población económicamente activa':[]})




for i in range(0,tabla1.shape[0],5):

    tabla_aux = tabla1.iloc[i:,0:]
    tabla_aux=tabla_aux.T
    tabla_aux.iloc[0,1]='Año'
    tabla_aux.iloc[1:,0]=tabla_aux.iloc[0,0]
    tabla_aux.iloc[0,0]='Departamento'
    tabla_aux=tabla_aux.rename(columns={i:'Departamento',i+1:'Año',i+2:'Población total',i+3:'Población en edad de trabajar',i+4:'Fuerza de trabajo'})
    tabla_aux=tabla_aux.drop(0)
    tabla=pd.concat([tabla,tabla_aux],axis=0)
    tabla.dropna(axis=1, inplace=True)
    

print(tabla.tail(5))

tabla.to_csv('D:/Colombia/Empleo/tabla2.csv',index=False,sep=';')
```
